# Gráficas:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset
df = pd.read_csv('resultados_deteccion.csv')

# Crear la gráfica de barras 
plt.figure(figsize=(10, 6))
sns.barplot(data=df[['Porcentaje Rojo', 'Porcentaje Verde']], palette=["red", "green"], ci=None)
plt.title('Comparación de Porcentajes Rojo y Verde')
plt.xlabel('Índice')
plt.ylabel('Porcentaje')
plt.show()

# Crear la gráfica de líneas
plt.figure(figsize=(10, 6))
sns.lineplot(data=df[['Porcentaje Rojo', 'Porcentaje Verde']], palette=["red", "green"])
plt.title('Tendencia de Porcentajes Rojo y Verde')
plt.xlabel('Índice')
plt.ylabel('Porcentaje')
plt.show()




```
